```python
import os
from astropy.io import fits
import numpy as np
import pandas as pd
from astropy.table import Table

from scipy import interpolate

from astropy import constants as const

```

```python
C = const.c.to('km/s')
C
```

```python
data=[]
count = 0
obsid = []
desig = []
C = 299792.458
dataSet= []
error_ = 0
for root, dirs, files in os.walk(r"H:\jupyter-code\ipython\Li\11-18及之后\142703条光谱"):
    print(root)  # 当前目录路径
    # print(files)  # 当前路径下所有非目录子文件

for file in files:
        hdu = fits.open(root + '/' + file)
        hdr = hdu[0].header#参数存在header里
        wave = hdu[0].data[2]
#         print(min(wave),max(wave))
#         print(len(wave))
#         break
        if  hdr['SNRR'] < 30:#筛选信噪比>30的
            hdu.close()
            continue

        

        z = hdu[0].header['Z']
        if abs(z)>=100:
            hdu.close()
            continue
        wavelength = hdu[0].data[2]
        # orgwavelength = [(i / (1 + z)) for i in wavelength]# 红移校正   无需校正

        orgflux = hdu[0].data[0]#读取流量flux，flux和wavelength存储在data中


        

        wavenew = np.linspace(3700, 9000, 5300)#插值，
        #波长，以及对应的流量
        f = interpolate.interp1d(wavelength, orgflux,fill_value="extrapolate", kind="slinear")#线性插值
        data = f(wavenew)#插值后的流量

        data = (data - np.min(data)) / (np.max(data) - np.min(data))#对流量归一化
               
       
        # desig obsid 用来跟sql下载的星表交叉，得到大气参数信息
        desig.append(hdr['DESIG'])
        
        
        data=np.append(data,hdr['OBSID'])
        hdu.close()

        
        
        dataSet.append(data)
        
        count+=1
        if count%100==0:
            print(count/len(files)*100,'%')
            dataSet=np.array(dataSet)
            desig = np.array(desig)
            desig = desig.reshape(-1,1)
            
            con = np.concatenate((dataSet,desig),axis=1)
            con = pd.DataFrame(con)
            con.to_csv(r'G:\高淇低分14万条全光谱数据.csv',mode='a',index=False,header=None)
            
            data=[]
            obsid = []
            desig = []
            dataSet= []

            
#最后剩下的部分
dataSet=np.array(dataSet)
desig = np.array(desig)
desig = desig.reshape(-1,1)

con = np.concatenate((dataSet,desig),axis=1)
con = pd.DataFrame(con)
con.to_csv(r'G:\高淇低分14万条全光谱数据.csv',mode='a',index=False,header=None)

# data=[]
# obsid = []
# desig = []
# dataSet= []        

    

```
