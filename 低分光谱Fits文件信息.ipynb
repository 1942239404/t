```python
data = {'wave_nonli':wavelength1[2514:2645],
        'flux_nonli':flux1[2514:2645],
        'wave_li':wavelength2[2514:2645],
        'flux_li':flux2[2514:2645]}
df = pd.DataFrame(data)
df
```

```python
df.to_csv('两个光谱的6600-6800波长处的流量值.csv',header=True,index=True,sep=',')
```

```python
def function(a, b):
    if a == b:
        return 1
    else:
        return 0
```

```python

 # 两条光谱波长取值相同
df['bool'] = df.apply(lambda x : function(x['wave_nonli'],x['wave_li']),axis = 1)
df
```

```python
count = 0
for i in df['bool']:
    if i == 1:
        count+=1
print(count)
```
